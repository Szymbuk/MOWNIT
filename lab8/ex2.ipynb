# a) $f(x) = x^3 - 5x$

- Metoda Newtona opiera się na wzorze iteracyjnym:
  $$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$$
- Dla funkcji $f(x) = x^3 - 5x$, jej pochodna wynosi $f'(x) = 3x^2 - 5$.
- Obliczenia krok po kroku dla punktu $x_0 = 1$:
  - **Iteracja 1**:
    - $f(1) = 1^3 - 5(1) = -4$
    - $f'(1) = 3(1)^2 - 5 = -2$
    - $x_1 = 1 - \frac{-4}{-2} = 1 - 2 = -1$
  - **Iteracja 2**:
    - $f(-1) = (-1)^3 - 5(-1) = 4$
    - $f'(-1) = 3(-1)^2 - 5 = -2$
    - $x_2 = -1 - \frac{4}{-2} = -1 + 2 = 1$
- **Wniosek**: Algorytm wpada w nieskończoną pętlę (oscylację) między wartościami $1$ a $-1$. Styczne w tych punktach kierują rozwiązanie z powrotem do poprzedniego punktu, co uniemożliwia zbieżność do pierwiastka.

## Wyznaczenie pierwiastków analitycznie

- Przyrównujemy funkcję do zera: $x^3 - 5x = 0$.
- Wyłączamy $x$ przed nawias: $x(x^2 - 5) = 0$.
- Otrzymujemy trzy pierwiastki:
  - $x_1 = 0$
  - $x_2 = \sqrt{5}$
  - $x_3 = -\sqrt{5}$

```python
import scipy.optimize as opt

def f_a(x):
    return x**3 - 5*x

def df_a(x):
    return 3*x**2 - 5

root_1 = opt.newton(f_a, x0=0.5, fprime=df_a)
root_2 = opt.newton(f_a, x0=2.5, fprime=df_a)
root_3 = opt.newton(f_a, x0=-2.5, fprime=df_a)

print(f"Pierwiastek 1: {root_1}")
print(f"Pierwiastek 2: {root_2}")
print(f"Pierwiastek 3: {root_3}")
```

### Alternatywa: Metoda bisekcji

```python
root_bisect = opt.bisect(f_a, a=1.5, b=3.0)
print(f"Pierwiastek (bisekcja): {root_bisect}")
```

# b) $f(x) = x^3 - 3x + 1$

- Mamy daną funkcję $f(x) = x^3 - 3x + 1$ oraz punkt startowy $x_0 = 1$.
- Pochodna funkcji to $f'(x) = 3x^2 - 3$.
- Podstawiając $x_0 = 1$ do wzoru na pochodną, otrzymujemy: $f'(1) = 3(1)^2 - 3 = 0$.
- Wzór metody Newtona zawiera dzielenie przez pochodną: $x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$.
- W pierwszej iteracji algorytm próbuje podzielić przez zero. Styczna do wykresu w punkcie $x = 1$ jest pozioma (znajduje się tam ekstremum lokalne), więc nie przecina osi OX.

## Rozwiązanie
- Wystarczy wybrać inny punkt startowy (np. $x_0 = 2$), w którym styczna nie jest pozioma.

```python
def f_b(x):
    return x**3 - 3*x + 1
def df_b(x):
    return 3*x**2 - 3

root_b = opt.newton(f_b, x0=2.0, fprime=df_b)
print(f"Znaleziony pierwiastek dla (b): {root_b}")
```


# c) $f(x) = 2 - x^5$

- Funkcja to $f(x) = 2 - x^5$, a punkt startowy to $x_0 = 0.01$.
- Pochodna wynosi $f'(x) = -5x^4$.
- Wartość pochodnej dla $x_0 = 0.01$ jest skrajnie bliska zeru: $f'(0.01) = -5 \cdot 10^{-8}$.
- Po podstawieniu do wzoru $x_1 = 0.01 - \frac{f(0.01)}{f'(0.01)}$, algorytm dzieli przez niezwykle małą liczbę.
- W rezultacie pierwsze przybliżenie jest "wyrzucane" gigantycznie daleko: $x_1 \approx 40\ 000\ 000$.
- Punkt początkowy $x_0 = 0.01$ leży zbyt blisko punktu przegięcia z poziomą styczną ($x=0$). Powoduje to, że algorytm osiągnie limit iteracji zanim zdąży "wrócić" w okolice rzeczywistego pierwiastka (który wynosi ok. $1.15$).

## Rozwiązanie
- Zmiana punktu startowego na oddalony od płaskiego rejonu funkcji, na przykład $x_0 = 1$.

```python
def f_c(x):
    return 2 - x**5
def df_c(x):
    return -5*x**4

root_c = opt.newton(f_c, x0=1.0, fprime=df_c)
print(f"Znaleziony pierwiastek dla (c): {root_c}")
```


# d) $f(x) = x^4 - 4.29x^2 - 5.29$

- Funkcja to $f(x) = x^4 - 4.29x^2 - 5.29$, z punktem początkowym $x_0 = 0.8$.
- Prawdziwe pierwiastki możemy obliczyć analitycznie (np. podstawiając $t = x^2$): wynoszą one dokładnie $x = 2.3$ oraz $x = -2.3$.
- Pochodna funkcji to $f'(x) = 4x^3 - 8.58x$.
- Jeśli przeprowadzimy kolejne iteracje dla $x_0 = 0.8$, otrzymamy: 
  - $x_1 \approx -0.7835$
  - $x_2 \approx 0.7892$
  - $x_3 \approx -0.7870$
  - $x_4 \approx 0.7878$
- Algorytm wpada w nieskończoną pętlę (tzw. cykl przyciągający). Wartości oscylują pomiędzy około $0.787$ a $-0.787$. Styczne w tych punktach na przemian wskazują na siebie, omijając docelowe pierwiastki.

## Rozwiązanie
- Należy uciec z pułapki cyklu, wybierając punkty startowe położone na zewnątrz od minimów lokalnych, na przykład $x_0 = 3$ dla prawego pierwiastka oraz $x_0 = -3$ dla lewego.

```python
def f_d(x):
    return x**4 - 4.29*x**2 - 5.29
def df_d(x):
    return 4*x**3 - 8.58*x
```

```python
x = 0.8
for i in range(5):
    fx = f_d(x)
    dfx = df_d(x)
    if dfx == 0: break
    x_next = x - fx/dfx
    print(f"Iter {i}: x={x:.4f}, f(x)={fx:.4f}, f'(x)={dfx:.4f}, x_next={x_next:.4f}")
    x = x_next
```

```python
root_d1 = opt.newton(f_d, x0=3.0, fprime=df_d)
root_d2 = opt.newton(f_d, x0=-3.0, fprime=df_d)
print(f"Znaleziony pierwiastek 1 dla (d): {root_d1}")
print(f"Znaleziony pierwiastek 2 dla (d): {root_d2}")
```
