```python
import numpy as np
from numpy.linalg import solve,norm
```

```python
f1 = lambda x1,x2: x1**2 + x2**2 -1
f2 = lambda x1,x2: x1**2 - x2
F_vector = np.array([f1,f2])
```

```python
a_00 = lambda x1,x2: 2*x1
a_01 = lambda x1,x2,: 2*x2
a_10 = lambda x1,x2: 2*x1
a_11 = lambda x1,x2: -1

J_func = np.array([[a_00,a_01],
              [a_10,a_11]])
```

```python
x_s = np.array([1.0,1.0])
N = 20
for i in range(N):
    J = np.array([ [J_func[0][0](x_s[0],x_s[1]), J_func[0][1](x_s[0],x_s[1])],
                   [J_func[1][0](x_s[0],x_s[1]), J_func[1][1](x_s[0],x_s[1])] ])

    F = np.array([F_vector[0](x_s[0],x_s[1]), F_vector[1](x_s[0],x_s[1])])

    sk = np.linalg.solve(J, -F)
    x_s = x_s + sk
print(x_s)
result1 = x_s
```

```python
x_s = np.array([1.0,1.0])
counter = 0
eps = 1e-14
while True:
    J = np.array([ [J_func[0][0](x_s[0],x_s[1]), J_func[0][1](x_s[0],x_s[1])],
                   [J_func[1][0](x_s[0],x_s[1]), J_func[1][1](x_s[0],x_s[1])] ])

    F = np.array([F_vector[0](x_s[0],x_s[1]), F_vector[1](x_s[0],x_s[1])])

    sk = np.linalg.solve(J, -F)
    new_x_s = x_s + sk
    if norm(sk)< eps:
        break
    x_s = new_x_s
    counter +=1
print(new_x_s)
result2 = new_x_s
```

```python
true_x = np.array([+np.sqrt(np.sqrt(5)/2 -1/2), np.sqrt(5)/2 - 1/2])

error1 = norm(result1-true_x)/norm(true_x)
error2 = norm(result2-true_x)/norm(true_x)

print(f'Błąd dla metody 1 (20 iteracji): {error1}')
print(f'Błąd dla metody 2 ({counter} iteracji): {error2}')
```
