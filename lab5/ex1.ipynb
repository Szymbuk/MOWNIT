```python
data = pd.read_csv("./USA_population.csv", delimiter=';')
x_s=  (data.Rok.values-1900)/40
y_s = data.Populacja.values
```

```python
polynomials = []
coefficients = []
for m in range(7):
    polynomial = np.polyfit(x_s,y_s,m)[::-1] # chcemy od najmniejszego wsp do największego
    coefficients.append(polynomial)
    #print(polynomial,"\n\n")
    fun = lambda x,p=polynomial: sum(c*pow(x,i) for i,c in enumerate(p))
    polynomials.append(fun)
polynomials = np.array(polynomials)
print(coefficients)
```

## a)

```python
TRUE_VALUE = 248_709_873
EXTRAPOLATED_YEAR = 1990
results = []
for fun in polynomials:
     v = fun((EXTRAPOLATED_YEAR-1900)/40)
     results.append(v)
results = np.array(results)

relative_error = abs((results-TRUE_VALUE)/TRUE_VALUE)
print("Błąd względny populacji h dla m = :")
for m in range(7):
    print(f'{m}: {round(relative_error[m-1],4)}')
```

## b)

```python
def AIC(y_true,y_estimated,k):
    n = len(y_true)
    return 2*k+ n*np.log(sum((y_true-y_estimated)**2)/n)

def AIC_c(y_true,y_estimated,k):
    n = len(y_true)
    return AIC(y_true,y_estimated,k) + (2*k*(k+1)/(n-k-1))
```

```python
# macierz każda kolumna to wartość estymacji dla konkretnego roku
# wierszami są wypisywane wyniki dla konkretnego wielomianu
population_estimation_by_year_by_polynomial =  []
for i in range(0,len(polynomials)):
    population_estimation_by_year_by_polynomial.append(polynomials[i](x_s))
    print(population_estimation_by_year_by_polynomial[i],"\n\n")
```

```python
Akaik_cryteria = []
for i in range(len(polynomials)):
    res = AIC_c(y_s,population_estimation_by_year_by_polynomial[i],len(coefficients[i]))
    Akaik_cryteria.append(round(res,4))
print(Akaik_cryteria)
```
