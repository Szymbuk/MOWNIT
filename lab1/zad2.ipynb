### zad2
a) $ \sqrt{x+1} -1, x \approx 0$

$\sqrt{x+1} - 1 = \frac{ (\sqrt{x+1} -1)*(\sqrt{x+1}+1)} { \sqrt{x+1} +1} = \frac{x+1-1}{\sqrt{x+1}+1} = \frac{x}{\sqrt{x+1}+1} $

```python
x_a = 0
basic_function_a = lambda x: np.sqrt(x+1)-1
optimized_function_a = lambda x: x/(np.sqrt(x+1)+1)
h_array = 10.0 ** -np.arange(0, 17)

basic_results_a =basic_function_a(h_array)
optimized_results_a = optimized_function_a(h_array)

non_relative_error_a = abs(basic_results_a - optimized_results_a)
relative_error_a = abs(basic_results_a - optimized_results_a) / abs(optimized_results_a)
```

```python
print(f'basic results: {basic_results_a}')
print(f'optimized results: {optimized_results_a}')
print(f'non relative_error: {non_relative_error_a}')
print(f'relative_error: {relative_error_a}')
```

```python
#plt.loglog(h_array, basic_results_a, 'ko-', label='błąd niezoptymalizowanej funkcji')
#plt.loglog(h_array, optimized_results_a, 'g--', label='błąd zoptymalizowanej funkcji')
plt.semilogx(h_array, relative_error_a, 'b--', label='unormowana różnica błędu między funkcjami')
plt.title("Zależność różnicy błędu w zależności od h")
plt.xlabel("h")
plt.ylabel("różnica błędu")
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.legend()
```

b) $x^2 - y^2, x \approx y$

$x^2 - y^2 = (x-y)*(x+y)$


```python
x_b = 1
basic_function_b = lambda x, h : (x+h)**2 - x**2
optimized_function_b = lambda x, h: (x-x+h)*(x+h +x)
# okazało się podczas testów, że istnieje znacząca różnica między działaniem x-x+h a x+h-x, to drugie dla małych h daje zero co uniemożliwia normalizację
h_array = 10.0 ** -np.arange(0, 17)

basic_results_b = basic_function_b(x_b,h_array)
optimized_results_b = optimized_function_b(x_b,h_array)

non_relative_error_b = abs(basic_results_b-optimized_results_b)
relative_error_b = abs(basic_results_b-optimized_results_b)/abs(optimized_results_b)

```

```python
print(f'basic results: {basic_results_b}')
print(f'optimized results: {optimized_results_b}')
print(f'non relative_error: {non_relative_error_b}')
print(f'relative_error: {relative_error_b}')
```

```python
plt.semilogx(h_array, relative_error_b, 'b--', label='unormowana różnica błędu między funkcjami')
plt.title("Zależność różnicy błędu w zależności od h")
plt.xlabel("h")
plt.ylabel("różnica błędu")
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.legend()
```
