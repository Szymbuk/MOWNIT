```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(67)
```

```python
def f(x):
    return 4 / (1 + x**2)
```

```python
exact_value = np.pi

n_values = np.logspace(1, 6, num=50, dtype=int)
```

```python
errors_mc = []

for n in n_values:
    x_random = np.random.uniform(0, 1, n)
    
    integral_mc = np.mean(f(x_random))
    
    rel_err = np.abs(integral_mc - exact_value) / exact_value
    errors_mc.append(rel_err)
```

```python

plt.figure(figsize=(10, 6))
plt.plot(n_values, errors_mc, 'o-', alpha=0.7, label='Metoda Monte Carlo')
theoretical_rate = errors_mc[0] * np.sqrt(n_values[0]) / np.sqrt(n_values)
plt.plot(n_values, theoretical_rate, 'r--', label='Tempo teoretyczne $1/\\sqrt{n}$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Liczba punktów próbkowania (n)')
plt.ylabel('Bezwzględna wartość błędu względnego')
plt.title('Błąd całkowania Monte Carlo w zależności od liczby punktów n')
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.7)
plt.show()
```
