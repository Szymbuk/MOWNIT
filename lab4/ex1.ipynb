```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def calculate_geometric_distance(points):
    n = len(points)
    distances = np.array([abs(points - p) for p in points])
    results = []
    for i in range(n):
        res = 1
        for j in range(n):
            if i==j:
                continue
            res *= distances[i][j]
        results.append(pow(res,1/(n-1)))
    #print(distances)
    #print(results)
    return results
```

```python
def draw_point_distance_plot(n = 5):
    fig, ax = plt.subplots()
    equal_distance_points = np.linspace(-1,1,n)
    #print(equal_distance_points)
    distances_equal = calculate_geometric_distance(equal_distance_points)
    ax.plot(equal_distance_points,distances_equal,color="red",label="punkty równoodległe")
    #print("\n\n\n")


    chebyshev_points = np.array([-np.cos(i*np.pi/(n-1)) for i in range(n)])
    #print(chebyshev_points)
    distances_chebyshev = calculate_geometric_distance(chebyshev_points)
    ax.plot(chebyshev_points,distances_chebyshev,color="blue",label="punkty Czebyszewa")
    #print("\n\n\n")


    coefficients = np.array([0]*n + [1])
    legendre_points = np.polynomial.legendre.legroots(coefficients)
    #print(legendre_points)
    distances_legendre = calculate_geometric_distance(legendre_points)
    ax.plot(legendre_points,distances_legendre,color="green",label="punkty Legendra")

    ax.set_ylabel("średnia geometryczna odległości od innych punktów")
    ax.set_xlabel("współrzędne punktu")
    ax.set_title("Średnia geometryczna odległości od pozostałych punktów w zależności od punktu")
    plt.legend()
    #print("\n\n\n")
```

```python
n_s = np.array([5,10,20])
for n in n_s:

    draw_point_distance_plot(n)
```
