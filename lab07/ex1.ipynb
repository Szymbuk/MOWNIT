$$v = \sqrt{\frac{2}{m} \int_{0}^{L} F \, dx}$$

```python
M = 0.075
integral_result = scipy.integrate.simpson(data.F,data.x)
result = np.sqrt(2/M * integral_result)
print(result)
```
