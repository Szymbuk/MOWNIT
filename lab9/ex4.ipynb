## Analiza Równania
Rozpatrywane równanie różniczkowe to typowe równanie liniowe pierwszego rzędu postaci $y' = \lambda y$, gdzie $\lambda = -5$.
Rozwiązanie ogólne takiego równania to $y(t) = C e^{\lambda t}$.
Uwzględniając warunek początkowy $y(0) = 1$, dokładne rozwiązanie analityczne przyjmuje postać:$$y(t) = e^{-5t}$$

## Stabilność Analityczna (Uwarunkowanie Problemu)
Stabilność analityczna problemu początkowego bada, jak małe zmiany (zaburzenia) warunku początkowego wpływają na rozwiązanie w miarę upływu czasu.
Załóżmy, że warunek początkowy został nieznacznie zaburzony o wartość $\epsilon$, czyli $\tilde{y}(0) = 1 + \epsilon$.
Rozwiązanie problemu zaburzonego to $\tilde{y}(t) = (1 + \epsilon)e^{-5t}$.
Błąd rozwiązania (różnica między rozwiązaniem zaburzonym a dokładnym) w dowolnej chwili $t$ wynosi:$$\delta(t) = \tilde{y}(t) - y(t) = (1 + \epsilon)e^{-5t} - e^{-5t} = \epsilon e^{-5t}$$Ponieważ $\lambda = -5 < 0$, błąd $\delta(t)$ dąży do zera, gdy czas $t \to \infty$:
$$\lim_{t \to \infty} \epsilon e^{-5t} = 0$$

## Wniosek
Rozwiązania powyższego równania są analitycznie stabilne (a dokładniej: asymptotycznie stabilne).Problem jest dobrze uwarunkowany, ponieważ dowolne zaburzenie warunku początkowego nie tylko nie rośnie z czasem, ale jest eksponencjalnie tłumione i z czasem zanika. Prawdziwe rozwiązanie naturalnie dąży do zera niezależnie od punktu startowego.

## Dowód Zbieżności Metody Eulera
Aby udowodnić zbieżność metody Eulera dla naszego problemu, musimy wyznaczyć jawną postać rozwiązania numerycznego $y_n$, a następnie obliczyć jego granicę przy kroku $h \to 0$.

### Krok 1: Sformułowanie wzoru metody Eulera
Rozpatrujemy równanie z części (a): $y' = -5y$ z warunkiem początkowym $y(0) = y_0 = 1$.
Wzór jawnej metody Eulera dla równania $y' = f(t, y)$ ma postać:$$y_{n+1} = y_n + h \cdot f(t_n, y_n)$$

### Krok 2: Zastosowanie metody do danego równania
Podstawiając naszą funkcję $f(t, y) = -5y$, otrzymujemy relację rekurencyjną:$$y_{n+1} = y_n + h(-5y_n)$$
$$y_{n+1} = y_n(1 - 5h)$$

### Krok 3: Wyznaczenie wyrazu ogólnego $y_n$
Rozwijając tę relację od warunku początkowego $y_0 = 1$, otrzymujemy:
- $y_1 = y_0(1 - 5h) = 1 \cdot (1 - 5h)$
- $y_2 = y_1(1 - 5h) = (1 - 5h) \cdot (1 - 5h) = (1 - 5h)^2$
- ... 
- $y_n = (1 - 5h)^n$

### Krok 4: Obliczenie granicy w punkcie $t$
Zgodnie z treścią zadania, chcemy zbadać granicę dla $t = nh$, przy $h \to 0$ i $n \to \infty$. Skoro $t = nh$, to możemy zapisać $n = \frac{t}{h}$.
Podstawiamy to do wzoru na $y_n$:$$y_n = (1 - 5h)^{\frac{t}{h}}$$
Teraz przekształcamy to wyrażenie, aby skorzystać ze wskazówki podanej w zadaniu, czyli definicji liczby $e$: $\lim_{x \to 0} (1 + x)^{\frac{1}{x}} = e$.
Zapiszmy wyrażenie w odpowiedniej formie, wprowadzając nową zmienną $x = -5h$. Gdy $h \to 0$, to również $x \to 0$:$$y_n = \left( (1 + (-5h))^{\frac{1}{-5h}} \right)^{-5t}$$
Przechodząc do granicy przy $h \to 0$:$$\lim_{h \to 0} y_n = \lim_{h \to 0} \left[ (1 + (-5h))^{\frac{1}{-5h}} \right]^{-5t}$$
Z podanej wskazówki wiemy, że wyrażenie w nawiasie kwadratowym dąży do $e$:$$\lim_{h \to 0} y_n = e^{-5t}$$

### Krok 5: Wniosek
Otrzymana granica ciągów numerycznych dla $h \to 0$ to dokładnie funkcja $e^{-5t}$.
Z części (a) wiemy, że dokładne rozwiązanie analityczne to $y(t) = e^{-5t}$.
Wykazaliśmy tym samym, że:$$\lim_{\substack{h \to 0 \\ n \to \infty \\ nh=t}} y_n = y(t)$$ co stanowi bezpośredni dowód na to, że jawna metoda Eulera jest w tym przypadku zbieżna.

## Stabilność Numeryczna Metody Eulera
W jawnej metodzie Eulera dla równania testowego postaci $y' = \lambda y$, kolejne wartości rozwiązania numerycznego wyrażają się wzorem:$$y_{n+1} = y_n + h\lambda y_n = y_n(1 + h\lambda)$$
Wyrażenie $R = 1 + h\lambda$ nazywamy czynnikiem wzmacniającym.
Aby metoda numeryczna była stabilna (błędy zaokrągleń i przybliżeń nie rosły w kolejnych iteracjach), moduł czynnika wzmacniającego musi być mniejszy lub równy 1:$$|1 + h\lambda| \le 1$$

## Zastosowanie do zadanego problemu
Z treści zadania wiemy, że $\lambda = -5$.
Użyty krok czasowy wynosi $h = 0.5$.
Obliczamy wartość czynnika wzmacniającego dla tych parametrów:$$R = 1 + 0.5 \cdot (-5) = 1 - 2.5 = -1.5$$
Sprawdzamy warunek stabilności:$$|R| = |-1.5| = 1.5$$
Widzimy wyraźnie, że $1.5 > 1$.

## Wniosek
Ponieważ moduł czynnika wzmacniającego jest ostro większy od 1, metoda Eulera z krokiem $h = 0.5$ dla tego równania jest numerycznie niestabilna.
Oznacza to, że w przeciwieństwie do analitycznego rozwiązania, które łagodnie dąży do zera, rozwiązanie numeryczne będzie zachowywać się patologicznie. Z racji ujemnego czynnika ($-1.5$), w każdym kroku wartość będzie zmieniać znak na przeciwny (oscylacje), a jej amplituda będzie wykładniczo rosnąć do nieskończoności ($y_n = (-1.5)^n$). Maksymalny krok $h$, dla którego ta metoda byłaby stabilna, to $h = \frac{2}{5} = 0.4$.

## Obliczenia Numeryczne dla $t = 0.5$

### Krok 1: Dane i szukane
Znamy warunek początkowy $y_0 = 1$ dla $t_0 = 0$, krok czasowy $h = 0.5$ oraz funkcję $f(t, y) = -5y$.
Chcemy obliczyć wartość przybliżoną $y_1$ dla chwili $t_1 = t_0 + h = 0.5$.
Ponieważ docelowy czas osiągamy po dokładnie jednym kroku, wystarczy wykonać tylko jedną iterację metody.

### Krok 2: Zastosowanie wzoru
Zgodnie z jawną metodą Eulera wzór wygląda następująco:$$y_1 = y_0 + h \cdot f(t_0, y_0)$$
Po podstawieniu naszej funkcji otrzymujemy postać:$$y_1 = y_0 - 5h y_0 = y_0(1 - 5h)$$

### Krok 3: Wykonanie obliczeń
Podstawiamy wartości liczbowe do wyprowadzonego wzoru:$$y_1 = 1 \cdot (1 - 5 \cdot 0.5)$$
$$y_1 = 1 - 2.5$$
$$y_1 = -1.5$$
## Wniosek
Wartość rozwiązania numerycznego w punkcie $t = 0.5$ wynosi $-1.5$.
Wynik ten doskonale ilustruje niestabilność zbadaną w punkcie (c) — podczas gdy dokładne rozwiązanie analityczne $y(0.5) = e^{-2.5} \approx 0.082$ jest dodatnie i bliskie zera, nasze numeryczne przybliżenie po zaledwie jednym kroku przyjęło dużą wartość ujemną.


```python
import math

def euler_step(f, t0, y0, h):
    y1 = y0 + h * f(t0, y0)
    return y1

def f(t, y):
    return -5 * y

t0 = 0.0
y0 = 1.0
h = 0.5

# t = 0.5 (wymaga tylko jednego kroku, bo t0 + h = 0.5)
t1 = t0 + h
y1 = euler_step(f, t0, y0, h)

print("--- Jawna Metoda Eulera ---")
print(f"Warunek początkowy: t0 = {t0}, y0 = {y0}")
print(f"Krok czasowy: h = {h}")
print(f"Obliczona wartość dla t = {t1}: y1 = {y1}")

y_exact = math.exp(-5 * t1)
print(f"Rozwiązanie analityczne dla t = {t1}: y(t) = {y_exact:.5f}")
```
